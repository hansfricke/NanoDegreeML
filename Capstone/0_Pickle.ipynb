#### Proposal data

```python
headers = ['id', 
            'teacher_id',
            'teacher_prefix',
            'state',
            'datetime',
            'grade_category',
            'subject_categories',
            'subject_subcategories',
            'title',
            'essay_1',
            'essay_2',
            'essay_3',
            'essay_4',
            'resource_summary',
            'previously_posted_projects',
            'approved']
dtypes = { 'id' : 'str', 
           'teacher_id' : 'str',
           'teacher_prefix' : 'str',
           'state' : 'str',
           'datetime' : 'str',
           'grade_category' : 'str',
           'subject_categories' : 'str',
           'subject_subcategories' : 'str',
           'title' : 'str',
           'essay_1' : 'str',
           'essay_2' : 'str',
           'essay_3' : 'str',
           'essay_4' : 'str',
           'resource_summary' :  'str',
           'previously_posted_projects' : 'int',
           'approved' : 'int'}
parse_dates = ['datetime']
proposal_data = pd.read_csv('train.csv', header = 0, na_values = 'NaN',
                    names=headers, dtype=dtypes, parse_dates=parse_dates)
```
